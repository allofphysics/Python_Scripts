```python
import pandas as pd
import numpy as np
from itertools import product
pd.set_option('display.max_rows',1000)
```

```python
ints=400
df=pd.DataFrame(np.array(range(ints)).reshape(20,20))
```

```python
indx_dct=df.to_dict()
```

```python
df
```

```python
def traversal_(r,c):
    dct={}
    bndry = df.shape
    #print bndry
    if (r >=0 and r< bndry[0]):
        dct[r]=[x for x in [c-1,c+1] if (x >=0 and x< bndry[1])]
        
    if (r-1 >=0 and r-1< bndry[0]): 
        dct[r-1]=[x for x in [c-1,c,c+1] if (x >=0 and x< bndry[1])]
    
    if (r+1 >=0 and r+1< bndry[0]):
        dct[r+1]=[x for x in [c-1,c,c+1] if (x >=0 and x< bndry[1])]
    
        
    return dct
```

```python
# from row 3 column 2 I can go to row 2 col 1,2 or 3
# or row 3 col 1 or 3 .... 
traversal_(1,1)
```

```python
df
```

```python
indicies=list(product(indx_dct.keys(),indx_dct.keys()))
```

```python
print indicies
```

```python
Range_df=pd.DataFrame({x:traversal_(x[0],x[1]) for x in indicies}).T.fillna('')
```
